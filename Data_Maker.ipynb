```python
import random
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
def single_path(up_bias,down_bias,intitial_value,total_ticks,min_volatility,max_volatility):
    a=intitial_value
    l=[a]
    up_down=[-1]*down_bias+[1]*up_bias
    for i in range(1,total_ticks):
        random.shuffle(up_down)
        a+=(up_down[0])*(random.randint(min_volatility,max_volatility))*(0.1)
        l.append(a)
    return np.array(l)
```

```python
l=single_path(50,100,100,200,100,150)
plt.plot(l)
plt.show()
```

```python
with open('data.dat','wb') as f:
    pickle.dump(l,f)
```
